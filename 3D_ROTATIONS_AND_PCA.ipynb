```
# Import dependencies

import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size':10})
```

```
fig=plt.figure(figsize=(15,15))
axes=[-1.5,1.5,-1.5,1.5,-1.5,1.5]
coef=.1
np.random.seed(12)
cube=np.random.rand(10000,3)
cube=np.c_[cube[:,0]*coef,cube[:,1:3]]
U,S,V_t=np.linalg.svd(cube)
polygon=cube.dot(V_t.T[:,:2])
inf=0
theta=0
while theta>=inf:
    if theta<20:
      rotation_matrix_x=np.array([[1,0,0],[0,np.cos(theta),-np.sin(theta)],[0,np.sin(theta),np.cos(theta)]])
      rotation_matrix_y = np.array([[np.cos(theta), 0, np.sin(theta)], [0, 1, 0], [ -np.sin(theta),0, np.cos(theta)]])
      rotation_matrix_z = np.array([[np.cos(theta),-np.sin(theta), 0], [np.sin(theta), np.cos(theta), 0], [0, 0 ,1]])
      axis_=np.linspace(-2,2,10)
      middle=np.zeros(10)
      R_x=cube.dot(rotation_matrix_x)
      R_y=cube.dot(rotation_matrix_y)
      R_z=cube.dot(rotation_matrix_z)
      ax1=fig.add_subplot(131,projection="3d")
      ax1.plot(R_x[:,0],R_x[:,1],R_x[:,2],"r.",alpha=.3)
      ax1.plot(axis_,middle,middle,'k.-',label="Axis")
      ax2 = fig.add_subplot(132, projection="3d")
      ax2.plot(R_y[:, 0], R_y[:, 1], R_y[:, 2], "g.",alpha=.3)
      ax2.plot(middle, axis_,middle, 'k.-',label="Axis")
      ax3 = fig.add_subplot(133, projection="3d")
      ax3.plot(R_z[:, 0], R_z[:, 1], R_z[:, 2], "b.",alpha=.3)
      ax3.plot(middle, middle, axis_,'k.-',label="Axis")
      ax1.set_title("$X$ $fixed$",color="r")
      ax2.set_title("$Y$ $fixed$",color="g")
      ax3.set_title("$Z$ $fixed$",color="b")
      ax1.set_xlim(axes[:2])
      ax1.set_ylim(axes[2:4])
      ax1.set_zlim(axes[4:6])
      ax1.set_xlabel("X",fontweight='bold')
      ax1.set_ylabel("Y",fontweight='bold')
      ax1.set_zlabel("Z",fontweight='bold')
      ax2.set_xlim(axes[:2])
      ax2.set_ylim(axes[2:4])
      ax2.set_zlim(axes[4:6])
      ax2.set_xlabel("X", fontweight='bold')
      ax2.set_ylabel("Y", fontweight='bold')
      ax2.set_zlabel("Z", fontweight='bold')
      ax3.set_xlim(axes[:2])
      ax3.set_ylim(axes[2:4])
      ax3.set_zlim(axes[4:6])
      ax3.set_xlabel("X", fontweight='bold')
      ax3.set_ylabel("Y", fontweight='bold')
      ax3.set_zlabel("Z", fontweight='bold')
      plt.suptitle("$3D$ $rotations$",fontsize=20,color="g")
      ax1.legend()
      ax2.legend()
      ax3.legend()
      plt.draw()
      plt.pause(.001)
      plt.clf()
    else:
      rotation_2d = -np.array([[np.cos(theta),-np.sin(theta)],[np.sin(theta),np.cos(theta)]])
      R_p=polygon.dot(rotation_2d)
      plt.plot(R_p[:,0],R_p[:,1],"c.",label="PCA polygon")
      plt.axis(axes[:4])
      plt.gca().set_xlabel("Component 1",fontweight="bold")
      plt.gca().set_ylabel("Component 2",fontweight="bold")
      plt.title("$2D$ $rotations$ $PCA$",fontsize=20,color="g")
      plt.draw()
      plt.pause(.2)
      plt.clf()
    theta+=1
```
